```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/canada_per_capita_income.csv")
print(df.head())
```

```python
plt.scatter(df.year, df["per capita income (US$)"], color="red", marker="o")
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada: Year vs Per Capita Income")
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["year"]], df["per capita income (US$)"])
```

```python
pred_2020 = reg.predict([[2020]])
print("Predicted per capita income (2020) [sklearn]:", pred_2020[0])
```

```python
print("Slope (m):", reg.coef_[0])
print("Intercept (b):", reg.intercept_)
```

```python
plt.scatter(df.year, df["per capita income (US$)"], color="red")
plt.plot(df.year, reg.predict(df[["year"]]), color="blue")
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Best Fit Line using sklearn")
plt.show()
```

```python
def fit_linear_regression(x, y):
    n = len(x)
    m = ( (n * np.sum(x*y)) - (np.sum(x)*np.sum(y)) ) / \
        ( (n * np.sum(x**2)) - (np.sum(x)**2) )
    b = (np.sum(y) - m*np.sum(x)) / n
    return m, b
```

```python
x = df["year"].values
y = df["per capita income (US$)"].values

m, b = fit_linear_regression(x, y)
print("Slope (m) [manual]:", m)
print("Intercept (b) [manual]:", b)
```

```python
pred_manual_2020 = m*2020 + b
print("Predicted per capita income (2020) [manual]:", pred_manual_2020)
```

```python
plt.scatter(x, y, color="red")
plt.plot(x, m*x + b, color="green")
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Best Fit Line using manual implementation")
plt.show()
```

```python
print("Predicted value for 2020 (sklearn):", pred_2020[0])
print("Predicted value for 2020 (manual):", pred_manual_2020)
print("Slope (m) [sklearn]:", reg.coef_[0], " | [manual]:", m)
print("Intercept (b) [sklearn]:", reg.intercept_, " | [manual]:", b)
```
