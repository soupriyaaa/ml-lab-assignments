```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df = pd.read_csv("/content/student_clustering.csv")
df.head()
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init="k-means++", random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Curve")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
centers = [(3, 3, 3), (0, 0, 0), (5, -5, 2)]
cluster_std = 1.2
```

```python
from sklearn.datasets import make_blobs
import numpy as np

X, y = make_blobs(
    n_samples=200,
    n_features=3,
    centers=centers,
    cluster_std=cluster_std,
    random_state=42
)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:,0], X[:,1], X[:,2], s=40)
ax.set_title("3D Scatter Plot of Generated Data")
plt.show()
```

```python
wcss_3d = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    wcss_3d.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wcss_3d, marker='o')
plt.title("Elbow Curve (3D Data)")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans_model = KMeans(n_clusters=3, random_state=42)
y_pred = kmeans_model.fit_predict(X)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:,0], X[:,1], X[:,2], c=y_pred, s=40, cmap='viridis')

ax.set_title("3D Clustering Result")
plt.show()
```
